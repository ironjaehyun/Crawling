```python
# stonesoupì¸ì§€ ì•„ë‹Œì§€ ë°˜ë“œì‹œ í™•ì¸ í•„ìš”
from bs4 import BeautifulSoup as bs
```

```python
# bsì—ê²Œ ë‚´ê°€ ê°€ì§€ê³ ìˆëŠ” í…ìŠ¤íŠ¸ ì •ë³´ë¥¼ ë³€í˜•í•´ì¤˜!
# bs("ë„˜ê²¨ë°›ì€ í…ìŠ¤íŠ¸", "íŒŒì‹±ë°©ë²•")
# * ë°˜ë“œì‹œ reqë¡œ ë°›ì•„ì˜¨ ë°ì´í„°ëŠ” bs ë³€í˜•ì´ í•„ìš”í•˜ë‹¤!
soup = bs(res1.text,"lxml")
```
