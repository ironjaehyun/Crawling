### BeautifulSoup
- req로 넘겨받은 텍스트 데이터를 컴퓨터가 이해할 수 있는 html형태로 변환

```python
# stonesoup인지 아닌지 반드시 확인 필요
from bs4 import BeautifulSoup as bs
```

```python
# bs에게 내가 가지고있는 텍스트 정보를 변형해줘!
# bs("넘겨받은 텍스트", "파싱방법")
# * 반드시 req로 받아온 데이터는 bs 변형이 필요하다!
soup = bs(res1.text,"lxml")
```

```python
# soup데이터 에서 선택해줘, 뉴스타이틀 태그만
# select("선택자") = 내가 필요한 태그만 수집이 가능
# select -> 무조건 복수개를 반환 -> 리스트
# select_one -> 무조건 한개만 반환 -> 복수개인 경우는 첫번째 데이터만 반환
news = soup.select("a.news_tit")
soup.select_one("a.news_tit")
```

```python
news[0].text
```

```python
# 사용자가 필요한 순수텍스트 정보를 담아주는 리스트 제작
# 데이터를 활용하기 위해서 (파일, 시각화 ...)
news_list = []
for i in news :
    news_list.append(i.text)
    print(i.text)
```
