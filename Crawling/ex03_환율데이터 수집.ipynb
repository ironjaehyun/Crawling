```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1, req를 활용하여 환율 사이트 정보 요청
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver")
```

```python
#2. bs데이터 변형
soup = bs(res.text,"lxml")
```

```python
#3. 통화이름 수집
money = soup.select("td.tit > a")
```
