```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'}
```

```python
# 1. req를 활용해서 멜론차트 사이트 요청
res = req.get("https://www.melon.com/chart/index.htm", headers = head)
```

```python
# 2. bs을 통해서 html데이터로 변형
soup = bs(res.text,"lxml")
```

```python
#3. 곡명 데이터를 수집을 해봅시다
# 클래스 값 중간에 공백이 있는 경우 !
# * 클래스가 복수개
# * 해결방법 = 공백을 지우고 .통해서 클래스를 이어줘라!
title = soup.select("div.ellipsis.rank01 > span > a")
```

```python
# 4. 가수이름 수집
# 데이터 개수를 반드시 검증해야한다. -> 정확한 개수가 있는 데이터를 수집
# 중복여부를 검사!
# 중복이 발생한 경우는 주변 태그들을 분석!
singer = soup.select("div.ellipsis.rank02 > span.checkEllipsis")
```
